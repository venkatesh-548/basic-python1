```python
#29 wap to check the number is armstrong or not

n=377
num=n
count=0
while n:
    count+=1
    n=n//10
rev=0
temp=num
while num:
    ld=num%10
    rev=rev+ld**count
    num//=10
if rev==temp:
    print("armstrong")
else:
    print("not an armstrong")

```

```python
# 30 wap to print the armstrong from mto n

m=1
n=1000
while m<=n:
    num=m
    temp=num
    ls=len(str(num))
    if int(ls)>2:
        sum=0
        while num:
            ld=num%10
            sum+=ld**ls
            num//=10
        if sum==temp:
            print(temp,end=" ")
    m+=1

    
```
